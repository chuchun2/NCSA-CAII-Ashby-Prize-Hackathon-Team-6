```python
r = np.zeros((len(y_name), len(x_name))) * np.nan
for itime in range(0, 133, 12):
    ds = xr.open_dataset("/home/chuchun2/hackathon_data/training.nc").sel(Time=itime)

    for iy in range(len(y_name)):
        y = ds[y_name[iy]].values.flatten()
        y_max = y.max()
        y_min = np.where(y == 0, y_max, y).min()
        y_new = np.where(y == 0, y_min, y)
        y2 = np.log(y_new)
        for ix in range(len(x_name)):
            x = ds[x_name[ix]].values.flatten()
            x_max = x.max()
            x_min = np.where(x == 0, x_max, x).min()
            x_new = np.where(x == 0, x_min, x)
            x2 = np.log(x_new)
            r[iy, ix] = np.absolute(np.corrcoef((x2-x2.mean())/x2.std(), (y2-y2.mean())/y2.std())[0, 1])
    
    fig = plt.figure(figsize=(25, 5))
    sns.heatmap(r, xticklabels=x_name, yticklabels=y_name, vmin=0, vmax=1, cmap="viridis", linewidths=.5)
    plt.title('t = '+str(itime)+' ('+str(ds['Times'].values)[2:-1]+')')
    plt.tight_layout()
    fig.savefig('./fig_norm_nonzero/abs_correlation_t'+str(itime)+'.png', dpi=150)
```
